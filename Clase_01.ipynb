Es un estndar utilizado en la mayora de los sistemas de gestin de bases de datos, como Microsoft SQL Server, Oracle, MySQL, PostgreSQL, entre otros.

SQL permite realizar acciones como crear, modificar y eliminar tablas, insertar, actualizar y eliminar datos, realizar consultas complejas y agrupar resultados, unir y relacionar tablas, entre otras acciones.

Es un lenguaje de alto nivel y fcil de aprender, que permite a los desarrolladores y administradores de bases de datos trabajar de manera eficiente y productiva con las bases de datos.
